```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\music_popularity_predictor\data\dataset.csv")
```

```python
df.dropna(inplace=True)
```

```python
# Plot popularity distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['popularity'], bins=30, kde=True)
plt.title("Distribution of Song Popularity")
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Correlation heatmap
numerical = df.select_dtypes(include='number')
plt.figure(figsize=(14, 10))
sns.heatmap(numerical.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Audio Features")
plt.show()
```

```python
# Popularity by genre (top 10 genres only)
top_genres = df['track_genre'].value_counts().head(10).index
subset = df[df['track_genre'].isin(top_genres)]
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x='track_genre', y='popularity', data=subset)
plt.xticks(rotation=45)
plt.title("Popularity Distribution by Top Genres")
plt.show()
```
